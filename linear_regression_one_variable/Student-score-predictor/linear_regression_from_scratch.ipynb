```python
import pandas as pd

df = pd.read_csv("student_scores.csv")
print(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Hours'], df['Scores'])
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.show()
```

```python
import numpy as np
def cost_function(x, y, w, b):
    m = len(x)
    f_wb = w * x + b 
    cost = (1 / (2 * m)) * np.sum((f_wb - y) ** 2)
    return cost 
```

```python
x = np.array(df['Hours'])
y = np.array(df['Scores'])
print(f'{x}, {y}')
```

```python
def gradient_calculation(x, y, w, b):
    m = len(x)
    df_dw = (1 / m) * np.sum((w * x + b - y) * x)
    df_db = (1 / m) * np.sum(w * x + b - y)
    return df_dw, df_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters):
    w = w_in
    b = b_in
    w_history = []
    b_history = []
    J_history = []
    for i in range(num_iters):
        df_dw, df_db = gradient_calculation(x, y, w, b)

        w = w - alpha * df_dw
        b = b - alpha * df_db
        J = cost_function(x, y, w, b)

        w_history.append(w)
        b_history.append(b)
        J_history.append(J)

    return w, b, J, w_history, b_history, J_history
```

```python
w = 10
b = 1.5
alpha = 0.05
iter = 1000
w_optimised, b_optimised, j_optimised, W, B, J= gradient_descent(x, y, w, b, alpha, iter)
for i in range(iter):
    print(f"{i} w: {W[i]}, b: {B[i]}, J: {J[i]}")
print(f"Optimised w: {w_optimised}, Optimised b: {b_optimised}, Final cost: {j_optimised}")
```

```python
def estimated(x):
    return w_optimised * x + b_optimised
```

```python

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color = 'blue', label='Actual Data')      # Scatter plot for 2D data
plt.plot(x, estimated(x), color='red', label='Estimated Line') # Line plot for estimation
plt.xlabel('X-Axis Title')
plt.ylabel('Y-Axis Title')
plt.title('Linear Regression Estimation')
plt.legend()
plt.show()
```

```python
x_test = 9.25
print(f"Predicted score for a student studying {x_test} hours: {estimated(x_test)}")
```
