```python
import pandas as pd

df = pd.read_csv("student_scores.csv")
```

```python
X = df[['Hours']]
y = df['Scores']
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Hours'], df['Scores'])
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
print(f'Weight = {model.coef_}, bias = {model.intercept_}')
```

```python
w = model.coef_[0]
b = model.intercept_
```

def estimated(x):
    return w * x + b

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Hours'], y, color = 'blue', label='Actual Data')      # Scatter plot for 2D data
plt.plot(df['Hours'], model.predict(X), color='red', label='Estimated Line') # Line plot for estimation
plt.xlabel('X-Axis Title')
plt.ylabel('Y-Axis Title')
plt.title('Linear Regression Estimation')
plt.legend()
plt.show()
```

```python
hours = 9.25

x_test = pd.DataFrame([[hours]], columns=['Hours'])

prediction = model.predict(x_test)[0]

print(f"Predicted score for a student studying {hours} hours: {prediction}")
```
